## Simulating a nonparametric inference (Ibrahim)

```python
import numpy as np
import pandas as pd
from scipy.stats import norm

enrollment_bootstrap = pd.read_csv("MC_Enrollment.csv")
enrollment_bootstrap['Gender'] = (enrollment_bootstrap['Gender'] == 'Female').astype(int) # Recoding as a binary outcome
```

```python
enrollment_bootstrap.head()
```

## Defining our Hypothesis:

We believe that the proortion of females is higher than the proportion of males at Montgomery College. Thus:

Ho: p_females = p_males

Ha: p_females > p_males

## Conducting our bootstrapping:

```python
np.random.seed(99)

# Checking our inital observed difference
observed_diff = np.mean(enrollment_bootstrap['Gender'])

# Creating an array to store bootstrap differences (using append function throughout sampling)
bootstrap_appends = []

# Performing bootstrap sampling
for i in range(10000):
    bootstrap_sample = np.random.choice(enrollment_bootstrap['Gender'], size=len(enrollment_bootstrap['Gender']), replace=True)
    bootstrap_diff = np.mean(bootstrap_sample)
    bootstrap_appends.append(bootstrap_diff)

# Calculating our p-value
p_value = np.mean(bootstrap_appends >= observed_diff) # "How many times is a difference in proportion like our observed difference likely to happen?"

print(p_value)
```

## Interpreting our p-value:

Our p-value was very large: 0.5078. Thus, we have no compelling evidence to reject the null hypothesis. We fail to reject the null hypothesis that the proportion of females at Montgomery College does not differ significantly from the proportion of males at Montgomery College.
